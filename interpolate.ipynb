```python
import numpy as np, math, matplotlib.pyplot as plt, scipy.ndimage as ndi

t = np.arange(0, 0.2, 1/44100)
s = sin(2 * math.pi * 400 * t)

s1 = s[0:100]
s2 = s[155:255]
```

```python
def smoothcat(x, y, z):
    print(x, y)
    li = len(x) - z
    ri = z
    # print(li, ri)
    left = x[li:len(x)]
    right = y[0:ri]
    sm = np.concatenate([left, right])
    print(sm)
    sm = ndi.gaussian_filter1d(sm, 3)
    print(sm)
    result = np.concatenate([x[0:li], sm, y[ri:len(y)]])
    print(result)
    return result
```

```python
test = np.concatenate([s1, s2])
```

```python
sc = smoothcat(s1, s2, 50)
plt.subplot(2, 1, 1)
plt.plot(np.arange(0, 200), test, 'o')
plt.subplot(2, 1, 2)
plt.plot(np.arange(0, 200), sc, 'o')
```
